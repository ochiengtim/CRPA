```python
import openai

def get_code_review(persona, code):
    """Please provide a code review in the tone described."""
    response = openai.ChatCompletion.create(
        model="gpt-4",  # Or gpt-3.5-turbo
        messages=[
            {"role": "system", "content": personalities[persona]},
            {"role": "user", "content": f"Review this code:\n{code}"}
        ]
    )
    return response.choices[0].message.content.strip()
```

```python
with gr.Blocks() as app:
    gr.Markdown("##  AI-Powered Code Review Personality Assistant (CRPA)")
    
    code_input = gr.Textbox(label="Paste your code", lines=12, placeholder="Enter Python code here...")
    persona_input = gr.Radio(["Mentor", "Drill Sergeant", "Buddy"], label="Choose Review Personality")
    review_output = gr.Textbox(label="Generated Review", lines=8)
    
    submit_button = gr.Button("Generate Review")
    submit_button.click(fn=generate_review, inputs=[code_input, persona_input], outputs=review_output)

app.launch(inline=True, share=False)  # Use inline=True to stay in Jupyter
```
